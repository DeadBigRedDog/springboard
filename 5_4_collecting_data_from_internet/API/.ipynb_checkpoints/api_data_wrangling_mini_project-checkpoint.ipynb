These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
#1.  Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017.

#I'm grabbing all of the data ordered by date asc, we could only pull in specific columns
url = "https://www.quandl.com/api/v3/datasets/FSE/EON_X?order=asc&start_date=2017-01-01&end_date=2017-12-31&api_key="+API_KEY
r = requests.get(url)

print(json.dumps(r.json(),sort_keys=True,indent=4))
```

```python
#2. Convert the returned JSON object into a Python dictionary.
raw_fse_afx_2017_data_dict = r.json()

#lets grab the data set that we need
fse_afx_2017_data_dict = raw_fse_afx_2017_data_dict['dataset']['data']
```

```python
#3.  Calculate what the highest and lowest opening prices were for the stock in this period.
highest = 0.00
lowest = 1000000.00

#lets unpack the tuples
for date,open_amt,high,low,close,change,traded_volume,turnover,last_price_of_day,daily_traded_units,daily_turnover in fse_afx_2017_data_dict:
    if open_amt > highest:
        highest = open_amt
    if open_amt < lowest:
        lowest = open_amt
        
print('Highest: ',highest)
print('Lowest: ',lowest)
```

```python
#4.  What was the largest change in any one day (based on High and Low price)?

largest_swing = 0.0

for date,open_amt,high,low,close,change,traded_volume,turnover,last_price_of_day,daily_traded_units,daily_turnover in fse_afx_2017_data_dict:
    swing = high - low
    if swing > largest_swing:
        largest_swing = swing
        
print('Largest 1 day swing: ', largest_swing)
```

```python
#5.  What was the largest change between any two days (based on Closing Price)?

previous_day_amt = 0.0
largest_2_day_swing = 0.0
date_of_swing = ''

#this is presuming the question is asking for consecutive days, not the largest swing between 2 days over the entire year
for date,open_amt,high,low,close,change,traded_volume,turnover,last_price_of_day,daily_traded_units,daily_turnover in fse_afx_2017_data_dict:
    #setting up the first day
    if(previous_day_amt == 0.0):
        previous_day_amt = close
    else:
        if largest_2_day_swing < abs(previous_day_amt - close):
            largest_2_day_swing = abs(previous_day_amt - close)
            date_of_swing = date
        #make sure the set the previous day amount for the next cycle
        previous_day_amt = close
        

highest = 0.00
highest_day = ''
lowest = 1000000.00
lowest_day = ''
largest_swing_between_2_days = 0.0

#this is presuming that the question is asking for any days difference over the entire year
for date,open_amt,high,low,close,change,traded_volume,turnover,last_price_of_day,daily_traded_units,daily_turnover in fse_afx_2017_data_dict:
    if close > highest:
        highest = open_amt
        highest_day = date
    if close < lowest:
        lowest = open_amt
        lowest_day = date

#based on the lowest and highest close prices will yeild the largest swing 
largest_swing_between_2_days = highest - lowest

print('Largest 2 day consecutive day swing: ', largest_2_day_swing)
print('Date: ',date_of_swing)
print('-----------------------------------------------------------')
print('Largest 2 day swing over the entire year: ',largest_swing_between_2_days )
print('Lowest Date: ',lowest_day)
print('Highest Date: ',highest_day)
```

```python
#6.  What was the average daily trading volume during this year?

traded_volume_total = 0.0
total_count = len(fse_afx_2017_data_dict)
average_daily_trading_volume = 0.0

for date,open_amt,high,low,close,change,traded_volume,turnover,last_price_of_day,daily_traded_units,daily_turnover in fse_afx_2017_data_dict:
    traded_volume_total += traded_volume
    
average_daily_trading_volume = traded_volume_total/total_count

print('Average Daily Trading Volume: ',average_daily_trading_volume)

```

```python
#7 (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

unsorted_trading_volume = []
median_amt = 0.0

#define custom median function.  This would be much easier using the statistics package.  :-)

def median(lst):
    quotient, remainder = divmod(len(lst), 2)
    if remainder:
        return sorted(lst)[quotient]
    return sum(sorted(lst)[quotient - 1:quotient + 1]) / 2.

for date,open_amt,high,low,close,change,traded_volume,turnover,last_price_of_day,daily_traded_units,daily_turnover in fse_afx_2017_data_dict:
    unsorted_trading_volume.append(traded_volume)

median_amt = median(unsorted_trading_volume)

print('Median Amount over entire year: ',median_amt)


```
